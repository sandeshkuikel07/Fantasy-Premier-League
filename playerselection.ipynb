```python
GAMEWEEK =38  #'''Provide Gameweek number here'''
import pandas as pd
import requests
import numpy as np

def get_captain_element(data):
    for entry in data:
        if entry.get('is_captain', False):
            return entry['element']
    return None

df = pd.read_csv('/content/drive/MyDrive/fpl_temp/Managers.csv')

GW=GAMEWEEK
for index, row in df.iterrows():
    manager = row['entry']
    url = 'https://fantasy.premierleague.com/api/entry/'+str(manager)+'/event/'+str(GW)+'/picks/'
    webpage = requests.get(url).json()
    points  = webpage['entry_history']['points']
    extra_transfer = webpage['entry_history']['event_transfers_cost']
    captain = get_captain_element(webpage['picks'])
    df.at[index,'captain'] = int(captain)
    df.at[index,'GW_P'] = points
    df.at[index,'E_T'] = -extra_transfer
    for i in range(0,11,1):
        player = webpage['picks'][i]['element']
        df.at[index,str(i)] = player
```

```python
import os

# Define the path to the directory
directory = '/content/drive/MyDrive/fpl_temp/' + str(GW)

# Create the directory if it does not exist
if not os.path.exists(directory):
    os.makedirs(directory)

# Define the full path to the file
file_path = os.path.join(directory, 'Player_selection_GW' + str(GW) + '.csv')

# Save the DataFrame to the specified path
df.to_csv(file_path, index=False)
```
